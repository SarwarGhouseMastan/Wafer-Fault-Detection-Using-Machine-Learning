```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import KNNImputer
```

```python
#Deleting unwanted columns
df.drop(columns=['Unnamed: 0','Unnamed: 0.1','wafer'], inplace=True)
```

```python
#Filling the NaN values with KNN Imputer and fitting, transforming the data
imputer = KNNImputer(n_neighbors=3, weights='uniform', missing_values=np.nan)
new_data = imputer.fit_transform(df)
```

```python
#Converting the new_data from Array to a DataFrame 
new_data = pd.DataFrame(new_data, columns=df.columns)
```

```python
#Saving the data in CSV format 
new_data.to_csv('clean_combined_data.csv')
```

```python
new_data.head()
```

```python
#Checking for NaN values after Imputing them with KNN Imputer
new_data.isnull().sum()
```

```python
# checking columns whihc have standard deviation of zero

col_to_drop = []
describe = df.describe()
for x in new_data.columns:
    if describe[x]['std']==0:
        col_to_drop.append(x)

col_to_drop
```

```python
describe
```

```python
#Just Creating a DataFrame to display the features whose SD is equal to zero
new_data.loc[:,col_to_drop]
```

```python
#Dropping all the columns whose SD is 0
new_data.drop(columns=col_to_drop, axis=1 , inplace=True)
```
